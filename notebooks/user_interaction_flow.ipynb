# üë§ User Interaction Flow for CNN Comparison Tool

This notebook outlines how users interact with the CNN model comparison tool, from model selection to result interpretation. This flow is designed to support both the current notebook-based prototype and a future web interface.

## 1Ô∏è‚É£ Model Selection

- User selects one of the available pre-trained models (e.g., ResNet18, UNet, DeepLabV3)
- Interface shows a dropdown or selection menu
- A short model description is displayed after selection (input size, task type)

## 2Ô∏è‚É£ Data Upload

- User uploads a `.tif` file (GeoTIFF) or chooses from example files
- Image is loaded and validated (format, bands, shape)
- Optionally, a preview of the image is displayed

## 3Ô∏è‚É£ Run Model Inference

- The selected model is applied to the input raster
- A progress bar or status message indicates inference is running
- The result is stored as a prediction mask or classified image

## 4Ô∏è‚É£ Visualize Results

- Input and prediction are shown side-by-side
- Color-coded map with class legend (if classification)
- Optional overlays and zoom features

## 5Ô∏è‚É£ Export / Compare Results

- Option to download output files (e.g. GeoTIFF, PNG)
- In future: enable selecting multiple models and compare their outputs visually
- Include performance metrics or logs if available

## üìä Summary in a Flowchart:

```python
from graphviz import Digraph

dot = Digraph()
dot.attr(rankdir="LR", size="12")

# Nodes with details
dot.node("A", "1- Model Selection\n- Choose from pretrained models\n- Show model info & requirements")
dot.node("B", "2- Upload GeoTIFF\n- Upload or use sample\n- Validate bands & format\n- Optional preview")
dot.node("C", "3- Run Inference\n- Apply model to image\n- Display progress\n- Generate prediction mask")
dot.node("D", "4- View Results\n- Side-by-side input/output\n- Show legend or classes\n- Highlight anomalies")
dot.node("E", "5- Export & Compare\n- Download results (GeoTIFF/PNG)\n- (Future) Compare multiple models\n- View logs/stats")

# Arrows between steps
dot.edge("A", "B", label="next")
dot.edge("B", "C", label="run")
dot.edge("C", "D", label="show")
dot.edge("D", "E", label="export")

dot.render("detailed_user_flow", format="png", cleanup=True)
dot
```
